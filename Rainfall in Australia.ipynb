```python
weather.drop(['Date', 'Location','WindDir3pm','WindDir9am','WindGustDir'],axis=1,inplace=True)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
#Converting string data to labels 
weather['RainToday']=label_encoder.fit_transform(weather['RainToday'])
```

```python
weather['RainTomorrow']=label_encoder.fit_transform(weather['RainTomorrow'])
```

#### Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X=weather.drop('RainTomorrow',axis=1)
y=weather['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=101)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)
```

```python
predict=lr.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(predict,y_test))
```

```python
print(confusion_matrix(predict,y_test))
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
print(np.sqrt(mean_squared_error(predict,y_test)))
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
prediction=dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(prediction,y_test))
```

```python
print(confusion_matrix(prediction,y_test))
```

```python
dtree.score(X_test,y_test)
```

#### Support Vector Machines

```python
from sklearn.svm import SVC
```

```python
model=SVC()
model.fit(X_train,y_train)
predict_svm=model.predict(X_test)
```

```python
print(classification_report(predict_svm,y_test))
```

```python
print(confusion_matrix(predict_svm,y_test))
```

```python
model.score(X_test,y_test)
```
